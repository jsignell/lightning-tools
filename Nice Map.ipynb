```python
from lightning import *
%matplotlib inline
```

```python
c = Region(city=cities['phoenix'])
step = .01
c.gridx = np.arange(-113.7, -109.5+step, step)
c.gridy = np.arange(31.5, 35.0+step, step)
```

```python
ds = c.get_ds(y='201*', m=8, func='grid')
```

```python
plt.figure(figsize=(12,8))
ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.set_extent([c.gridx.min(), c.gridx.max(), c.gridy.min(), c.gridy.max()])

ax = background(ax)
ax = dem(ax)
urban(ax)
c.plot_grid(c.FC_grid,zorder=10, alpha=.5, vmin=10, vmax=60, cbar=True, ax=ax);
```

```python
plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection=ccrs.PlateCarree())
ax.set_extent([c.gridx.min()+1, c.gridx.max()-1, c.gridy.min()+1, c.gridy.max()-1])

ax = background(ax)

# this is the new way but it is bad in couple ways. Namely that it has a max tile n of 9
ax.add_raster(SRTM3Source(), cmap=plt.get_cmap('Greys', 10))
urban(ax, edgecolor='red', facecolor='None')
c.plot_grid(c.FC_grid,zorder=10, alpha=.5, vmin=10, vmax=60, cbar=True, ax=ax);
```
